# Simple Linear Regression
-----

## Step 1: <span style="text-decoration:underline">Importing the libraries</span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: <span style="text-decoration:underline">Importing the dataset</span>

```python
dataset = pd.read_csv("Salary_Data.csv")
```

```python
dataset.head()
```

## Step 3:  <span style="text-decoration:underline">Data analysis</span>

```python
# To analyze descriptive statistics of data
dataset.describe()
```

Here, we can notice Salary ranges from 37731 to 122391 and a median of 65237.


We can additionally visualize the data distribution using Seaborn's histplot

```python
plt.title('Salary Distribution Plot')
sns.histplot(dataset['Salary'])
plt.show()
```

Then we check the relationship between Salary and Experience -

```python
# Relationship between Salary and Experience
plt.scatter(dataset['YearsExperience'], dataset['Salary'], color = 'lightcoral')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

## Step 4: <span style="text-decoration:underline">Split the dataset into dependent/independent variables</span>

```python
# independent variables
x = dataset.iloc[:,:-1].values
# dependent variables
y = dataset.iloc[:,-1].values
```

## Step 5:  <span style="text-decoration:underline">Spliting the dataset into thetraining set and test set</span>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,
                                                   random_state=0)
```

## Step 6: <span style="text-decoration:underline">Training the simple Linear Regression model on training set</span>

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Step 7: <span style="text-decoration:underline">Predicting The Test set results</span>

```python
y_pred = regressor.predict(x_test)
```

## Step 8:  <span style="text-decoration:underline">Visualsing the Training set results</span>

```python
plt.scatter(x_train, y_train, color="red")
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Experience (Training set)")
plt.xlabel("years of Experience")
plt.ylabel("salary")
plt.show()
```

## Step 9: <span style="text-decoration:underline">Visualsing the Test set resultsm</span>

```python
plt.scatter(x_test, y_test, color="red")
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Experience (Test set)")
plt.xlabel("years of Experience")
plt.ylabel("salary")
plt.show()
```

```python
# Making a single prediction (for example the salary of an employee with 12 years of experience)
print(regressor.predict([[12]]))
```

## step 10: <span style="text-decoration:underline">The coefficients</span>

```python
from sklearn.metrics import mean_squared_error, r2_score
print(f"Coefficients: {regressor.coef_}")
# The mean squared error
print(f"Mean squared error: {mean_squared_error(y_test, y_pred):.2f}")
# The coefficient of determination: 1 is perfect prediction
print(f"Coefficient of determination: {r2_score(y_test, y_pred):.2f}")
```
